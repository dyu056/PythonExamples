From the derivation submitted via PDF, the following equation obtains the coordinate of L4 and L5
$$L_4 = (r_{1,2}/2 - \pi_2 r_{12},\sqrt{3}r_{12}/2)$$
$$L_5 = (r_{1,2}/2 - \pi_2 r_{12},-\sqrt{3r_{12}}/2)$$

```python
# Constants
m1 = 5.972e24; #Earth's mass
m2 = 7.34767309e22; #Moon's mass
r_12 = 384400000; #Distance between the Earth and the Moon.
pi_1 = m1/(m1 + m2);
pi_2 = m2/(m1 + m2);
L4_x = r_12/2 - pi_2*r_12;
L5_x = L4_x;
L4_y = np.sqrt(3)*r_12/2;
L5_y = -np.sqrt(3)*r_12/2;

#The following prints L4 and L5 in sequence, unit in meters.
print(L4_x)
print(L4_y)
print(L5_x)
print(L5_y)
```

Assuming y = 0, derivation on pdf is conducted to obtain the following expression for the rest of the three Lagrange point
$$0 = -\frac{\mu}{r_{12}^3}+\frac{\mu_1(x+\pi_2r_{12})}{|x+\pi_2r_{12}|^3}+\frac{\mu_2(x-\pi_1r_{12})}{|x-\pi_1r_{12}|^3} = f(x)$$
Therefore, for Lagrange point, one can plot the above function and find the root by checking where the function intersect at $f(x) = 0$

For $X = xr_{12}$, we can simplify the above expression further, detailed process is on PDF I submitted
$$f(X) = -X+\frac{\pi_1(X+\pi_2)}{|X+\pi_2|^3}+\frac{\pi_2(X-\pi_1)}{|X-\pi_1|^3}$$

```python
X = np.linspace(-1.5, 2, num=1000)
f = -X + pi_1*(X+pi_2)/(np.abs(X+pi_2)**3)+pi_2*(X-pi_1)/(np.abs(X-pi_1)**3);
plt.plot(X,f)
plt.ylim(top=2)  # adjust the top leaving bottom unchanged
plt.ylim(bottom=-2)  # adjust the bottom leaving top unchanged
```

```python
# Next step is to find the index where f = 0
fat0_index = np.where((f > -0.01) & (f < 0.01));
print(fat0_index)
#The following X then yields the Lagrange point
Xlag = X[fat0_index];
print(Xlag)
#The following array then contains the x coordinate of L1,L2 and L3, y = 0 for these lagrange points
xlag = Xlag*r_12;
print(xlag)
```
