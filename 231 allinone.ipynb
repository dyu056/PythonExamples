```python
def diffract(theta:list,w,s,N,lmbda,scale):
    k = 2*pi/lmbda;
    v = 0.5*k*s*np.sin(theta);
    u = 0.5*k*w*np.sin(theta);
    y = w**2 * (((sin(u))/(u))*((sin(N*v))/(sin(v))))**2
    y = y * scale
    return y
```
